## Exploratory Data Analysis

Exploratory Data Analysis is an approach to analyse data to summarise their main characteristics and often with visual methods.

### Univariate Data Analysis

The term univariate analysis refers to the analysis of one variable. You can remember this because the prefix “uni” means “one.”

The purpose of univariate analysis is to understand the distribution of values for a single variable.

#### 1. Histogram

A frequency distribution shows how often each different value in a set of data occurs. A histogram is the most commonly used graph to show frequency distributions.

A histogram divides up the range of possible values in a data set into classes or groups. For each group, a rectangle is constructed with a base length equal to the range of values in that specific group and a length equal to the number of observations falling into that group ( x-axis describes the groups and the y-axis describes the frequency of occurrence).

More on histograms on this Matplotlib [Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)


```python
#Histrogram showing Bill Length
x=penguins['bill_depth_mm']
#when you are plotting the histogram,plt.hist(have the necessary parameters here)
plt.hist(x, bins=10, histtype='bar')
#when to give my plot a title
plt.title('Histogram showing Bill length')
plt.xlabel('Bill Length')
plt.ylabel('Frequency of values')
plt.show()

```

```python
#
plt.hist(penguins['bill_depth_mm'])
plt.title('Histogram showing Bill Depth')
plt.xlabel('Bill depth')
plt.ylabel('Frequency')
plt.show()
```

#### 2. Countplot


A countplot is a type of categorical plot commonly used in data visualization. It displays the number of occurrences of each category in a dataset. It is a useful tool for analyzing the distribution and frequency of categorical variables.


Countplot [Documentation](https://seaborn.pydata.org/generated/seaborn.countplot.html)

```python
# Plotting the Distribution of the Species Variable
ax = sns.countplot(data = penguins, x='species')
ax.set_title('Count Distribution of Species Variable');
```

**Observation**

*The Adelie species made a the majority of the Penguins in the dataset while the Chinstrap were the minority.*

```python
# Plotting the Distribution of the Island Variable
sns.set(rc={'figure.figsize':(10, 6)}) #adjusting the width and height as desired
ax=sns.countplot(x='island',data=penguins)
ax.set_title('Count Distribution of the Island Variable');
#Biscoe Island has the highest count followed by Dream.
```

**Observation**

*The Biscoe Island had the most Penguins while the Torgersen island had the least number of penguins.*

```python
# Plotting the Distribution of the Sex Variable
#sns.set(rc={'figure.figsize':(8, 4)}) #adjusting the width and height as desired
ax = sns.countplot(x='sex',data=penguins)
ax.set_title('Count Distribution of the Sex Variable');
```

```python
penguins['sex'].value_counts()
```

**Observation**

*Majority of the Penguins were male.*

#### 3. Boxplot

A boxplot displays the five-number summary of a set of data.
The five-number summary is the minimum, first quartile, median, third quartile, and maximum.

Box plots give a good graphical image of the concentration of the data. They also show how far the extreme values are from most of the data.

download.png

```python
#Checking for anomalies and outliers in the Age of the Respondent Variable
fig, ax = plt.subplots(2,2, figsize=[20,20])
sns.boxplot(x='bill_length_mm', data=penguins, ax=ax[0,0])
sns.boxplot(x= 'bill_depth_mm', data=penguins, ax=ax[0,1])
sns.boxplot(x='flipper_length_mm', data=penguins, ax=ax[1,0])
sns.boxplot(x= 'body_mass_g', data=penguins, ax=ax[1,1])

# Setting the Title of the Plots
ax[0,0].set_title("Distribution of the Bill Length Variable")
ax[0,1].set_title("Distribution of the Bill Depth Variable")
ax[1,0].set_title("Distribution of the Flipper Length Variable")
ax[1,1].set_title("Distribution of the Body Mass Variable");
```

### Bivariate Analysis

Bivariate Analysis is where you are comparing two variables to study their relationship.

There should be a y_value and and x_values

#### 1. Scatter Plot

A scatter plot (aka scatter chart, scatter graph) is a type of visualization that displays relationship between two numeric variables.

It uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point.

Scatter Plot [Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

```python
# Distribution of the Bill Length and Body Mass Variables
#sns.set(rc={'figure.figsize':(15, 6)}) #adjusting the width and height as desired
ax = sns.scatterplot(x='bill_length_mm',y='body_mass_g',data=penguins)
ax.set_title("Bill Length against Body Mass")
plt.show()
```

```python
#Plotting distribution of the Bill Length and Body Mass Variables using matplotlib
#plt.figure(figsize=(15,6)) #adjusting the width and height as desired
plt.scatter(data = penguins,x='bill_length_mm',y='body_mass_g')
plt.title("Bill Length against Body Mass")
plt.show()
```

**Observation**

 *There is a positive,moderate linear relation*

```python
# Distribution of the Bill Length and the Bill Depth Variables
#plt.figure(figsize=(15, 6)) # adjusting the width and height as desired using matplotlib library
ax = sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=penguins)
ax.set_title("Bill Length against Bill Depth");
#negative,moderate linear relation.
```

#### 2. Barplots

A barplot (or barchart) is one of the most common types of graphic. It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar.
The size of the bar represents its numeric value.

Bargraph Matplotlib [Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)

```python
# Distribution of Species Against Body Mass
#plt.figure(figsize=(15, 6))
ax = sns.barplot(x='species',y='body_mass_g',data=penguins)
ax.set_title("Species against Body Mass");
#
```

```python
# Distribution of sex Against Body Mass
#plt.figure(figsize=(15, 6))
ax = sns.barplot(x='sex',y='body_mass_g',data=penguins)
ax.set_title("Sex against Body Mass");
```

####4. Boxplot and countplot

Both Boxplot and Countplot are versatile visualization tools that can be used on Univariate and Bivariate analysis as shown below,depending on how the data is structured and the specific purpose of the analysis.

Boxplot and Countplot are used primarily for Univariate Analysis.



```python
# Distribution of the Island Variable Against Body Mass
plt.figure(figsize=(15, 6))
ax = sns.boxplot(x='island',y='body_mass_g',data=penguins)
ax.set_title("Island against Body Mass");
```

```python
#Distribution of both sex and Island
plt.figure(figsize=(15, 5))
ax=sns.countplot(x='island',hue='sex', data=penguins)
ax.set_title('Count Distribution of the Island Variable against Sex');
```

**Difference between Bar graph and Histogram**

The histogram is a term that refers to a graphical representation that shows data by way of bars to display the frequency of numerical data while The bar graph is a graphical representation of data that uses bars to compare different categories of data.

Histogram display the frequency of occurrences while Bar graph compares different categories of data.
